# 5. **Data Analysis**

```python
# Rank players by defending stats.

top_10_defenders = data.sort_values(by='defending', ascending=False).head(10)

print(top_10_defenders[['name', 'ova', 'heading accuracy', 'interceptions', 'composure', 'marking', 'standing tackle', 'sliding tackle']])
```

```python
top_10_scorers = data.sort_values(by='attacking', ascending=False).head(10)

print(top_10_scorers[['attacking', 'name', 'ova', 'finishing', 'volleys', 'power', 'shot power', 'long shots', 'mentality', 'penalties']])
```

```python
top_10_dribblers =  data.sort_values(by='dribbling', ascending=False).head(10)

print(top_10_dribblers[['dribbling', 'name', 'ova', 'acceleration', 'skill', 'ball control', 'sprint speed', 'agility']])
```

```python
top_10_vision = data.sort_values(by='vision', ascending=False).head(10)
print(top_10_vision[['vision', 'name', 'ova', 'short passing', 'long passing', 'positioning']])
```
